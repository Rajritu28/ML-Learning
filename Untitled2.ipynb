```python
def isMissing(x):
    missing_total = x.isnull().sum().sort_values(ascending=False)
    percent = (x.isnull().sum() * 100/x.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([missing_total, percent], axis=1, keys=['Total', 'Percent'])
    print(missing_data[missing_data['Percent'] > 0]['Percent'])
    missing_data[missing_data['Percent'] > 0]['Percent'].plot.bar(by = 'Percent', figsize=(6,4))
```

```python
def income_cat_proportions(data,col):
    return data[col].value_counts() / len(data)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(df,'Sex'),
    "Train": income_cat_proportions(train_set,'Sex'),
    "Test": income_cat_proportions(test_set,'Sex'),
}).sort_index()
compare_props["Train Set %diff"] = 100 * compare_props["Train"] / compare_props["Overall"] - 100
compare_props["Test Set %diff"] = 100 * compare_props["Test"] / compare_props["Overall"] - 100
```
